```python
# Modules you need
import numpy as np, matplotlib.pyplot as plt, pandas as pd

# The Line fitting m:slope, b:x-intercept
def line(x, m, b):
    return m * x + b

# Colors
c = ["y", "orange", "r", "m", "b"]
# VGS values
vsg = [2, 3, 4, 5, 6]


# You must modify this value according to your result #####################
# VDS(sat) vector
vds_sat=1
##########################################################################

# Reading the data points
data = pd.read_csv("data2.csv", on_bad_lines="skip")
plt.figure(figsize=(8,6))
for i in range(0, 5):
    vds, Ids = data["vds%d" % (i + 1)], data["ids%d" % (i + 1)]
    vds, Ids = vds[:].to_numpy(dtype="float"), Ids[:].to_numpy(dtype="float")
    # Converting Current to [A]
    Ids = Ids * 0.001
    # Fitting
    m, b = np.polyfit(vds[vds > vds_sat], Ids[vds > vds_sat], 1)
    # Early Voltage Va = b/m (the distance from origin to x-intercept)
    va = b / m
    lambda_val = 1 / va
    # Ploting the fit and Exp. Data (only the acceptable range of function)
    fit_x = np.linspace(-va * 1.1, max(vds), 500)
    fit_y = line(fit_x, m, b)
    plt.plot(vds, Ids, "o",color=c[i], label="$V_{GS}$=%dV Exp."%(vsg[i]), alpha=0.6,ms=5)
    plt.plot(fit_x, fit_y, "--",color=c[i], label="$V_{GS}$=%dV fit"%(vsg[i]),lw=1)
    
    # placing a text with values on your graph. 
    plt.text(-240, (7 - i * 0.5) * 1e-3, "$V_A$=%.2fV" % (va), fontsize=14, color=c[i])
    plt.text(-180, (7 - i * 0.5) * 1e-3,"$\\lambda$=%.4f" % (lambda_val),fontsize=14,color=c[i],)

# plot axis setting
plt.xlim(-250, 12)
plt.xticks([-250,-200,-150,-100,-50,0,12],fontsize=14)
plt.ylim(0, 10e-3)
plt.yticks(1e-3 * np.arange(0, 12, 2), ["0", "2", "4", "6", "8", "10"],fontsize=14)

plt.xlabel("$V_{ds} \\; [V]$",fontsize=14)
plt.ylabel("$I_{ds} \\; [mA]$",fontsize=14)
plt.legend(ncol=5, fontsize=8)

# Save your result:
plt.savefig('CD4007(VA).svg')

plt.show()
```
